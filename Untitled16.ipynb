```python
# Load YOLO model
def load_yolo_model():
    net = cv2.dnn.readNet("yolov3.weights", "yolov3.cfg")
    layer_names = net.getLayerNames()
    output_layers = [layer_names[i[0] - 1] for i in net.getUnconnectedOutLayers()]
    return net, output_layers

net, output_layers = load_yolo_model()
```

```python
def detect_objects_yolo(frame, net, output_layers):
    height, width, _ = frame.shape
    blob = cv2.dnn.blobFromImage(frame, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
    net.setInput(blob)
    outs = net.forward(output_layers)

    class_ids = []
    confidences = []
    boxes = []
    for out in outs:
        for detection in out:
            for obj in detection:
                scores = obj[5:]
                class_id = np.argmax(scores)
                confidence = scores[class_id]
                if confidence > 0.5:  # confidence threshold
                    center_x = int(obj[0] * width)
                    center_y = int(obj[1] * height)
                    w = int(obj[2] * width)
                    h = int(obj[3] * height)
                    x = int(center_x - w / 2)
                    y = int(center_y - h / 2)
                    boxes.append([x, y, w, h])
                    confidences.append(float(confidence))
                    class_ids.append(class_id)

    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)
    return [(boxes[i[0]], confidences[i[0]]) for i in indexes]

def draw_detected_boxes(frame, detections):
    for box, conf in detections:
        x, y, w, h = box
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
    return frame

frames = extract_frames(video_path)
detections_per_frame = []
for frame in frames:
    detections = detect_objects_yolo(frame, net, output_layers)
    frame_with_boxes = draw_detected_boxes(frame, detections)
    detections_per_frame.append(frame_with_boxes)

print(f'Detected objects in frames.')
```

```python
import pytesseract

def extract_text_from_boxes(frame, boxes):
    texts = []
    for box, conf in boxes:
        x, y, w, h = box
        plate_img = frame[y:y+h, x:x+w]
        text = pytesseract.image_to_string(plate_img, config='--psm 8')  # --psm 8 assumes a single line of text
        texts.append(text.strip())
    return texts

# Extract text from detected license plates
texts_per_frame = []
for frame in detections_per_frame:
    boxes = detect_objects_yolo(frame, net, output_layers)  # detect boxes again for accurate regions
    texts = extract_text_from_boxes(frame, boxes)
    texts_per_frame.append(texts)

print('Extracted text from license plates.')
```

```python
import matplotlib.pyplot as plt

def show_frame_with_text(frame, texts):
    plt.figure(figsize=(10, 10))
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.title('\n'.join(texts))
    plt.axis('off')
    plt.show()

# Show the first frame and its detected texts
show_frame_with_text(detections_per_frame[0], texts_per_frame[0])
```
